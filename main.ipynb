```python

import os
from pathlib import Path

class FileManager:

    APP_NAME = "ReportingDCG"

    @staticmethod
    def get_cache_dir():
        base = Path(os.getenv("LOCALAPPDATA")) / FileManager.APP_NAME
        cache = base / "cache"
        cache.mkdir(parents=True, exist_ok=True)
        return cache

    @staticmethod
    def save_csv(df, filename):
        file_path = FileManager.get_cache_dir() / filename
        df.to_csv(file_path, index=False)
        return file_path
```
